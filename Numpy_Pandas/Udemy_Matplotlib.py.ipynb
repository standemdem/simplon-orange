# Matplotlib

## Line Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(0,10,10)
y = np.sin(x)

plt.plot(x,y)
plt.show() # in case the graph doesn't show up immediatly
```

```python
plt.plot(x,y)
plt.xlabel("Time") # adds x label
plt.ylabel("Some function of time") # adds y label
plt.title("My cool chart") # adds chart title
plt.show()
```

```python
x = np.linspace(0,10,100) # we add 10 times more points to make it more curvy
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

## Scatterplot

What we try to do in machine learning:
We have a bunch of datas, but we don't know the exact function where it came from.
There can also be noise added to that function.

=> what we really are trying to do is FUNCTION APPROXIMATION

```python
A = pd.read_csv('data_1d.csv', header = None).values
x= A[:,0]
y= A[:,1]

plt.scatter(x,y) # to create a cloud of dots
plt.show()
```

```python
x_line = np.linspace(0,100,100)
y_line = 2*x_line +1

plt.scatter(x,y)
plt.plot(x_line, y_line)
plt.show()
```

## Histogram

It is a frequency count in each bucket of values

```python
plt.hist(x)
plt.show()
```

```python
R = np.random.random(10000)
#we verify if the random function is truly a balanced random function
plt.hist(R)
plt.show()
```

We can also control the number of buckets of values that we add to the histogram

```python
plt.hist(R, bins=20)
plt.show()
```

One assumption in linear regression is that the error is normally distributed, and we can check it by doing a histogram

```python
y_actual = 2*x + 1
residuals = y - y_actual

plt.hist(residuals)
plt.show()
```

## Plotting Images

An image is just a matrix of numbers

```python
df = pd.read_csv('train.csv')
df.shape
```

```python
im = im.reshape(28,28)

plt.imshow(im)
```

```python
plt.imshow(im, cmap="gray")
plt.show()
```
