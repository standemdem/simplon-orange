### The apply Function !

If i want to assign a new column value where eache cell is derived from the values already in tis row

For ex: X2 => X1*X2

df['x1x2'] = df.apply(lambda row: row['x1'] * row['x2'], axis =1)

We pass in axis =1 so the function gets applied across each row instead of each column
It's kind of like a map function

```python
from datetime import datetime

datetime.strptime("1949-05","%Y-%m")
```

```python
df['dt'] = df.apply(lambda row: datetime.strptime(row['month'],"%Y-%m"),axis=1)
df.info()
```
